## Data Exploration and Cleaning

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../data/raw/train.csv')
```

```python
df.head()
```

```python
pd.options.display.max_columns=300
pd.options.display.max_rows=300
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe(include='all')
```

```python
df.isna().sum()
```

```python
df_clean = df.copy()
```

```python
threshold = 0.2 * len(df)
```

```python
columns_to_drop = df.columns[df.isnull().sum() > threshold]
```

```python
df_clean = df_clean.drop(columns=columns_to_drop)
```

```python
df_clean.info()
```

```python
num_cols = df_clean.select_dtypes(include=['float64', 'int64']).columns
for col in num_cols:
    mode_value = df_clean[col].mode()[0]
    df_clean[col] = df_clean[col].fillna(mode_value)
```

```python
cat_cols = df_clean.select_dtypes(include=['object']).columns
df_clean = df_clean.drop(columns=cat_cols)
```

```python
df_clean.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 16))
sns.heatmap(df_clean[num_cols].corr()[['drafted']], annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.show()
```

```python
df_cleaned = df_clean.copy()
```

## Assess Baseline

```python
y = df_cleaned.pop('drafted')
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
df_scaled = scaler.fit_transform(df_cleaned)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test,  y_train, y_test = train_test_split(df_scaled, y, test_size=0.2, random_state=42, stratify=y)
```

```python
pred_value = y_train.mode()
```

```python
y_base = np.full((len(y_train), 1), pred_value)
```

```python
from sklearn.metrics import accuracy_score as acs
from sklearn.metrics import f1_score as f1s
```

```python
print(acs(y_base, y_train))
print(f1s(y_base, y_train, average="weighted"))
```

## Initial Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
```

```python
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
rf_classifier.fit(X_train, y_train)
```

```python
y_train_preds = rf_classifier.predict(X_train)
y_val_preds = rf_classifier.predict(X_test)
```

```python
accuracy = acs(y_test, y_val_preds)
conf_matrix = confusion_matrix(y_test, y_val_preds)
report = classification_report(y_test, y_val_preds)

print(f"Accuracy: {accuracy:.4f}")
print(f"Confusion Matrix:\n{conf_matrix}")
print(f"Classification Report:\n{report}")
```

```python
print(acs(y_train_preds, y_train))
print(f1s(y_train_preds, y_train, average="weighted"))
```

```python
print(acs(y_val_preds, y_test))
print(f1s(y_val_preds, y_test, average="weighted"))
```

```python
from sklearn.metrics import roc_auc_score, roc_curve, auc

# Get the probability scores for the test set
y_prob = rf_classifier.predict_proba(X_test)[:, 1]  # Probability estimates for the positive class

# Compute the ROC AUC score
roc_auc = roc_auc_score(y_test, y_prob)
print(f"ROC AUC Score: {roc_auc:.4f}")

# Compute ROC curve
fpr, tpr, thresholds = roc_curve(y_test, y_prob)
roc_auc = auc(fpr, tpr)

# Plot ROC curve
plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc='lower right')
plt.show()
```

## Grid Search using stratified Kfold for Hyperparamater Tuning

```python
from sklearn.model_selection import GridSearchCV, StratifiedKFold

rfgs = RandomForestClassifier(random_state=42)
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
```

```python
param_grid = {
    'n_estimators': [100, 200, 500],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 5]
}
```

```python
grid_search = GridSearchCV(
    estimator=rfgs,
    param_grid=param_grid,
    scoring='roc_auc',  # Optimize for ROC AUC
    cv=skf,             # Use StratifiedKFold for cross-validation
    n_jobs=-1,          # Use all available CPU cores
    verbose=2           # Verbosity level
)
```

```python
grid_search.fit(X_train, y_train)
```

```python
print("Best hyperparameters:", grid_search.best_params_)
print("Best ROC AUC score:", grid_search.best_score_)
```
