```python
performance_var = ['Ortg','usg', 'eFG', 'TS_per', 'ORB_per', 'DRB_per', 'AST_per','TO_per', 'FTM','FTA',
                   'FT_per', 'twoPM', 'twoPA', 'twoP_per', 'TPM', 'TPA', 'TP_per', 'blk_per', 'stl_per', 
                   'ftr', 'adjoe', 'pfr', 'rimmade', 'rimmade_rimmiss', 'midmade', 'midmade_midmiss',
                   'dunksmade', 'dunksmiss_dunksmade', 'stops', 'bpm', 'obpm', 'dbpm', 'gbpm', 'mp', 'dgbpm']

# from group_12_krml_adv_mla_2024.features import feature_thresholding as fth
# threshold = fth.IQRThresholding()

# for col in performance_var:
#    df_copy = threshold.using_iqr('training', df_copy, col)

processor = []
def iqr_clipper(data,col, q1=0.25, q3=0.75):

    Q1 = data[col].quantile(q1)
    Q3 = data[col].quantile(q3)

    IQR = Q3-Q1

    lower_bound = Q1-1.5*IQR
    upper_bound = Q3+1.5*IQR
    
    data[col] = data[col].apply(lambda x: max(lower_bound, min(x, upper_bound)))

    processor.append({col:[{"lower_bound":lower_bound, "upper_bound":upper_bound}]})
    return data
```

```python
for col in performance_var:
    df_copy = iqr_clipper(df_copy, col)
```

### Modeling and Evaluation

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import roc_auc_score, roc_curve, auc, confusion_matrix, precision_score, recall_score, classification_report

from sklearn.ensemble import RandomForestClassifier


def train_and_evaluate_model(X, y, model, kf):
    """
    Train and evaluate the model using Stratified K-Fold Cross-Validation.

    Args:
        X: Features matrix.
        y: Target vector.
        model: The machine learning model to train.
        kf: StratifiedKFold instance for cross-validation.

    Returns:
        A dictionary containing ROC AUC scores, precision scores, recall scores,
        confusion matrices, and classification reports for each fold.
    """
    # Lists to store results
    results = {
        'roc_auc_scores': [],
        'precision_scores': [],
        'recall_scores': [],
        'confusion_matrices': [],
        'classification_reports': []
    }
    

    for fold, (train_index, val_index) in enumerate(kf.split(X, y), 1):
        X_train, X_val = X[train_index], X[val_index]
        y_train, y_val = y.iloc[train_index], y.iloc[val_index]

        # Train the model
        model.fit(X_train, y_train)
        
        # Predict probability scores for the positive class
        y_prob = model.predict_proba(X_val)[:, 1]
        y_pred = model.predict(X_val)  # Predictions for confusion matrix, precision, and recall

        # Compute ROC AUC score
        score = roc_auc_score(y_val, y_prob)
        results['roc_auc_scores'].append(score)

        # Compute confusion matrix, precision, recall, and classification report
        cm = confusion_matrix(y_val, y_pred)
        precision = precision_score(y_val, y_pred)
        recall = recall_score(y_val, y_pred)
        report = classification_report(y_val, y_pred, output_dict=True)

        # Store metrics
        results['confusion_matrices'].append(cm)
        results['precision_scores'].append(precision)
        results['recall_scores'].append(recall)
        results['classification_reports'].append(report)

        # Plot ROC curve
        fpr, tpr, _ = roc_curve(y_val, y_prob)
        plt.plot(fpr, tpr, alpha=0.3, label=f'ROC fold {fold} (area = {auc(fpr, tpr):.2f})')

    return results

def plot_roc_curve():
    """
    Plot the ROC curve across all folds.
    """
    plt.plot([0, 1], [0, 1], linestyle='--', lw=2, color='navy', alpha=0.8)
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve Across All Folds')
    plt.legend(loc='lower right')
    plt.show()

def print_evaluation_summary(results):
    """
    Print the evaluation summary including ROC AUC, precision, recall,
    confusion matrices, and classification reports.

    Args:
        results: A dictionary containing evaluation metrics for each fold.
    """
    mean_roc_auc = np.mean(results['roc_auc_scores'])
    std_roc_auc = np.std(results['roc_auc_scores'])
    mean_precision = np.mean(results['precision_scores'])
    std_precision = np.std(results['precision_scores'])
    mean_recall = np.mean(results['recall_scores'])
    std_recall = np.std(results['recall_scores'])

    print(f"Mean ROC AUC Score from Cross-Validation: {mean_roc_auc:.4f}")
    print(f"Standard Deviation of ROC AUC Scores: {std_roc_auc:.4f}")
    print(f"Mean Precision Score from Cross-Validation: {mean_precision:.4f}")
    print(f"Standard Deviation of Precision Scores: {std_precision:.4f}")
    print(f"Mean Recall Score from Cross-Validation: {mean_recall:.4f}")
    print(f"Standard Deviation of Recall Scores: {std_recall:.4f}\n")

    for fold, (cm, report) in enumerate(zip(results['confusion_matrices'], results['classification_reports']), 1):
        print(f"Confusion Matrix for fold {fold}:\n{cm}\n")
        # print(f"Classification Report for fold {fold}:\n{classification_report(report)}\n")

def plot_feature_importance(model, X):
    """
    Plot feature importances of the trained model.

    Args:
        model: The trained machine learning model.
        X: Features matrix.
    """
    # Get feature importances
    feature_importances = model.feature_importances_
    feature_names = X.columns if hasattr(X, 'columns') else np.arange(X.shape[1])
    
    # Create a DataFrame for feature importances
    importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': feature_importances})
    importance_df = importance_df.sort_values(by='Importance', ascending=False)

    # Plot feature importances
    plt.figure(figsize=(10, 6))
    plt.barh(importance_df['Feature'], importance_df['Importance'], color='skyblue')
    plt.xlabel('Importance')
    plt.title('Feature Importances')
    plt.gca().invert_yaxis()  # Highest importance at the top
    plt.show()

# Initialize Stratified K-Folds
kf = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)

# Initialize the Random Forest model
rf_model = RandomForestClassifier(random_state=42)

# Train and evaluate the model
results = train_and_evaluate_model(X_scaled, y, rf_model, kf)

# Plot ROC curve
plot_roc_curve()

# Print evaluation summary
print_evaluation_summary(results)

# # Plot feature importances
# plot_feature_importance(xgb_model, X)
```

### Tuning Hyperparameters

```python
from sklearn.model_selection import StratifiedKFold, GridSearchCV

def tune_hyperparameters(model, param_grid, X, y):
    """
    Tune hyperparameters using GridSearchCV.

    Args:
        X: Features matrix.
        y: Target vector.

    Returns:
        The best model found by GridSearchCV.
    """
    # 5 fold stratifiedkfold
    skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

    #  GridSearchCV
    grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=skf,
                               scoring='roc_auc', verbose=1, n_jobs=-1)

    # Fit GridSearchCV
    grid_search.fit(X, y)

    # Print the best parameters and score
    print(f"Best Parameters: {grid_search.best_params_}")
    print(f"Best ROC AUC Score: {grid_search.best_score_:.4f}")

    # Return the best model
    return grid_search.best_estimator_
```

```python
# Define the parameter grid to search
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [10, 20, 30],
    'min_samples_split': [5, 10],
    'min_samples_leaf': [2, 4]
    # 'max_features': ['auto', 'sqrt', 'log2']
}

best_model = tune_hyperparameters(rf_model, param_grid, X_scaled, y)
```
