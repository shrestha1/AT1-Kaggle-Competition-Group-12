```python
import torch
import torch.nn as nn
import torch.optim as optim
```

```python
class SimpleNN(nn.Module):
    def __init__(self, input_dim):
        super(SimpleNN, self).__init__()
        self.fc1 = nn.Linear(input_dim, 128)
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(128, 64)
        self.relu = nn.ReLU()
        self.fc3 = nn.Linear(64, 32)
        self.relu = nn.ReLU()
        self.fc4 = nn.Linear(32, 16)
        self.relu = nn.ReLU()
        self.fc5 = nn.Linear(16, 1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        x = self.fc1(x)
        x = self.relu(x)
        x = self.fc2(x)
        x = self.relu(x)
        x = self.fc3(x)
        x = self.relu(x)
        x = self.fc4(x)
        x = self.relu(x)
        x = self.fc5(x)
        x = self.sigmoid(x)
        return x.squeeze(1)  # Squeeze the output to have shape (batch_size,)
```

```python
from skorch import NeuralNetClassifier

# Assuming X_train is your training data with shape (n_samples, n_features)
input_dim = X_train.shape[1]

# Define the Skorch NeuralNetClassifier
net = NeuralNetClassifier(
    module=SimpleNN,           # Your defined neural network class
    module__input_dim=input_dim,  # Pass the input dimension
    criterion = nn.BCELoss,      # Binary Cross-Entropy Loss for binary classification
    max_epochs=50,             # Number of epochs to train
    lr=0.0001,                   # Learning rate
    optimizer=torch.optim.Adam, # Optimizer
    optimizer__weight_decay=1e-5,  # L2 regularization (weight decay)
    iterator_train__shuffle=True, # Shuffle training data
)
```

```python
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
y_train = y_train.astype('float32')
y_test = y_test.astype('float32')
```

```python
net.fit(X_train, y_train)
```
