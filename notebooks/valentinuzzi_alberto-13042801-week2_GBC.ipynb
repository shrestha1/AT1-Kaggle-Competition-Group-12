## Data Exploration and Cleaning

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../data/raw/train.csv')
```

```python
df_clean = df.copy()
```

```python
threshold = 0.2 * len(df)
```

```python
columns_to_drop = df.columns[df.isnull().sum() > threshold]
```

```python
df_clean = df_clean.drop(columns=columns_to_drop)
```

```python
df_clean.info()
```

```python
num_cols = df_clean.select_dtypes(include=['float64', 'int64']).columns
for col in num_cols:
    df_clean[col] = df_clean[col].fillna(df_clean[col].median())
```

```python
cat_cols = df_clean.select_dtypes(include=['object']).columns
```

```python
df_clean[cat_cols].head()
```

```python
df_clean[cat_cols].describe()
```

```python
df_clean = df_clean.drop(columns=cat_cols)
```

```python
df_clean.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 16))
sns.heatmap(df_clean[num_cols].corr()[['drafted']], annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.show()
```

```python
import seaborn as sns

corr_matrix = df_clean.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=False, cmap='coolwarm', fmt=".2f", square=True)
plt.title('Correlation Matrix Heatmap')
plt.figure(figsize=(20, 15))
plt.show()
```

Remove highly correlated features

```python
correlation_matrix = df_clean.corr().abs()

upper_tri = correlation_matrix.where(
    np.triu(np.ones(correlation_matrix.shape), k=1).astype(bool)
)

threshold = 0.9

to_drop = [column for column in upper_tri.columns if any(upper_tri[column] > threshold)]

df_clean = df_clean.drop(columns=to_drop)
```

```python
df_clean.info()
```

```python
df_cleaned = df_clean.copy()
```

## Grid Search using stratified Kfold for Hyperparamater Tuning

```python
from sklearn.model_selection import GridSearchCV, StratifiedKFold

gbgs = GradientBoostingClassifier(random_state=42)
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
```

```python
param_grid = {
    'n_estimators': [50, 100, 200],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 4, 5],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
}            
```

```python
grid_search = GridSearchCV(
    estimator=gbgs,
    param_grid=param_grid,
    scoring='roc_auc',  # Optimize for ROC AUC
    cv=skf,             # Use StratifiedKFold for cross-validation
    n_jobs=-1,          # Use all available CPU cores
    verbose=2           # Verbosity level
)
```

```python
grid_search.fit(X_trained, y_trained)
```

```python
print("Best hyperparameters:", grid_search.best_params_)
print("Best ROC AUC score:", grid_search.best_score_)
```
