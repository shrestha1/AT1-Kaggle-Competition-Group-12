## Data Exploration and Cleaning

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../data/raw/train.csv')
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df_clean = df.copy()
```

```python
threshold = 0.2 * len(df)
```

```python
columns_to_drop = df.columns[df.isnull().sum() > threshold]
```

```python
df_clean = df_clean.drop(columns=columns_to_drop)
```

```python
df_clean.info()
```

```python
num_cols = df_clean.select_dtypes(include=['float64', 'int64']).columns
for col in num_cols:
    mode_value = df_clean[col].mode()[0]
    df_clean[col] = df_clean[col].fillna(mode_value)
```

```python
cat_cols = df_clean.select_dtypes(include=['object']).columns
df_clean = df_clean.drop(columns=cat_cols)
```

```python
df_clean.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 16))
sns.heatmap(df_clean[num_cols].corr()[['drafted']], annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.show()
```

```python
df_cleaned = df_clean.copy()
```

## Initial Random Forest

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
gb_classifier = GradientBoostingClassifier(n_estimators=100, random_state=42)
```

```python
gb_classifier.fit(X_train, y_train)
```

```python
y_train_preds = gb_classifier.predict(X_train)
y_val_preds = gb_classifier.predict(X_test)
```

```python
from group_12_krml_adv_mla_2024.models.performance import print_classifier_outcomes
print_classifier_outcomes(y_val_preds, y_test, average="weighted")
```

```python
print_classifier_scores(y_preds=y_train_preds, y_actuals=y_train, average="weighted", set_name='Training')
```

```python
print_classifier_scores(y_preds=y_val_preds, y_actuals=y_test, average="weighted", set_name='Testing')
```

```python
from group_12_krml_adv_mla_2024.models.performance import compute_roc_curve

fpr, tpr, roc_auc = compute_roc_curve(gb_classifier, X_test, y_test, positive_class=1)

# Plot ROC curve
plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc='lower right')
plt.show()
```

## Grid Search using stratified Kfold for Hyperparamater Tuning

```python
from sklearn.model_selection import GridSearchCV, StratifiedKFold

gbgs = GradientBoostingClassifier(random_state=42)
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
```

```python
param_grid = {
    'n_estimators': [50, 100, 200],      
    'learning_rate': [0.01, 0.1, 0.2],    
    'max_depth': [3, 4, 5],               
    'min_samples_split': [2, 5, 10],      
    'min_samples_leaf': [1, 2, 4],        
    'subsample': [0.8, 1.0]   
}            
```

```python
grid_search = GridSearchCV(
    estimator=gbgs,
    param_grid=param_grid,
    scoring='roc_auc',  # Optimize for ROC AUC
    cv=skf,             # Use StratifiedKFold for cross-validation
    n_jobs=-1,          # Use all available CPU cores
    verbose=2           # Verbosity level
)
```

```python
grid_search.fit(X_train, y_train)
```

```python
print("Best hyperparameters:", grid_search.best_params_)
print("Best ROC AUC score:", grid_search.best_score_)
```
