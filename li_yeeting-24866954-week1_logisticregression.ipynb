```python
#Count missing values
num2=df_test.select_dtypes('number').columns

cnt = df_train[num].isna().sum()
per = ((cnt/df_train.shape[0])*100).round(4)

cnt2 = df_test[num2].isna().sum()
per2 = ((cnt2/df_test.shape[0])*100).round(4)

missing_cnt = pd.concat([per, per2], axis=1, keys=['Missing_Percent__train', 'Missing_Percent__test'])
missing_cnt.sort_values(['Missing_Percent__train', 'Missing_Percent__test'], ascending=False)
```

**Fill the missing values with zero**

```python
df_train['pick'] = df_train['pick'].fillna(0)
df_test['pick'] = df_test['pick'].fillna(0)

df_train['Rec_Rank'] = df_train['Rec_Rank'].fillna(0)
df_test['Rec_Rank'] = df_test['Rec_Rank'].fillna(0)
```

**Create binary features for Pick and Rec_Rank**

```python
df_train['Picked'] = (df_train['pick'] > 0).astype(int)
df_test['Picked'] = (df_test['pick'] > 0).astype(int)

df_train['Recruited'] = (df_train['Rec_Rank'] > 0).astype(int)
df_test['Recruited'] = (df_test['Rec_Rank'] > 0).astype(int)
```

```python
from sklearn.metrics import roc_auc_score, roc_curve, auc
```

```python
#Probabilities
y_train_pred = lr.predict_proba(X_train)[:, 1]
print("Train - AUROC:\n", roc_auc_score(y_train, y_train_pred))

y_val_pred = lr.predict_proba(X_val)[:, 1]
print("Val - AUROC:\n", roc_auc_score(y_val, y_val_pred))
```

**Plot the AUROC curve**

```python
fpr, tpr, thresholds = roc_curve(y_val, y_val_pred) 
roc_auc = auc(fpr, tpr)
plt.figure()  
plt.plot(fpr, tpr, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], 'k--', label='Worst case')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve for NBA Draft')
plt.legend()
plt.show()
```

* The model seems to be too good to use. Moving forward, feature engineering and dimensionality reduction will be conducted to distinguish the most significant predictors and address any overfitting issues.
